```python
import numpy as np 
import matplotlib.pyplot as plt 

np.random.seed(42)
```

```python
#  x 
x = np.linspace(0, 2, 1000)

# 1k 65%135%1
def generate_k_scenario1(n_samples=1000):
    # k
    k = np.where(
        np.random.random(n_samples) < 0.65,
        np.random.uniform(1, 2, n_samples),  # 65%  1-2 
        np.random.uniform(0, 1, n_samples)   # 35%  0-1 
    )
    return k

# 2k010
def generate_k_scenario2(n_samples=1000):
    # 51.5
    k = np.random.normal(5, 1.5, n_samples)
    # 0-10
    k = np.clip(k, 0, 10)
    return k

# 
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# 1
k_samples1 = generate_k_scenario1(100)
for k in k_samples1[:10]:  # 10
    ax1.plot(x, x**k, alpha=0.3)
ax1.set_title('1: k 65%>1, 35%<1')
ax1.set_xlabel('x')
ax1.set_ylabel('x^k')
ax1.grid(True)
ax1.set_ylim(0, 4)

# 2
k_samples2 = generate_k_scenario2(100)
for k in k_samples2[:10]:  # 10
    ax2.plot(x, x**k, alpha=0.3)
ax2.set_title('2: k0-10')
ax2.set_xlabel('x')
ax2.set_ylabel('x^k')
ax2.grid(True)
ax2.set_ylim(0, 4)

# k
ax1_hist = ax1.inset_axes([0.65, 0.65, 0.3, 0.3])
ax1_hist.hist(k_samples1, bins=20)
ax1_hist.set_title('k value distribution')
ax1_hist.set_xlim(0, 2)

ax2_hist = ax2.inset_axes([0.65, 0.65, 0.3, 0.3])
ax2_hist.hist(k_samples2, bins=20)
ax2_hist.set_title('k value distribution')
ax2_hist.set_xlim(0, 10)

plt.tight_layout()
plt.show()
```
