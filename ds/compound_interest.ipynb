```python
import numpy as np 
import matplotlib.pyplot as plt 

np.random.seed(42)
```

```python
# 创建 x 值序列
x = np.linspace(0, 2, 1000)

# 场景1：k值 65%概率大于1，35%概率小于1
def generate_k_scenario1(n_samples=1000):
    # 使用随机数来生成符合要求的k值
    k = np.where(
        np.random.random(n_samples) < 0.65,
        np.random.uniform(1, 2, n_samples),  # 65% 概率落在 1-2 之间
        np.random.uniform(0, 1, n_samples)   # 35% 概率落在 0-1 之间
    )
    return k

# 场景2：k值在0到10之间呈现正态分布
def generate_k_scenario2(n_samples=1000):
    # 使用正态分布，均值为5，标准差为1.5
    k = np.random.normal(5, 1.5, n_samples)
    # 将值限制在0-10之间
    k = np.clip(k, 0, 10)
    return k

# 创建子图
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# 场景1的可视化
k_samples1 = generate_k_scenario1(100)
for k in k_samples1[:10]:  # 只画10条线以避免过于密集
    ax1.plot(x, x**k, alpha=0.3)
ax1.set_title('场景1: k值 65%>1, 35%<1')
ax1.set_xlabel('x')
ax1.set_ylabel('x^k')
ax1.grid(True)
ax1.set_ylim(0, 4)

# 场景2的可视化
k_samples2 = generate_k_scenario2(100)
for k in k_samples2[:10]:  # 只画10条线以避免过于密集
    ax2.plot(x, x**k, alpha=0.3)
ax2.set_title('场景2: k值在0-10间呈正态分布')
ax2.set_xlabel('x')
ax2.set_ylabel('x^k')
ax2.grid(True)
ax2.set_ylim(0, 4)

# 添加k值分布的直方图
ax1_hist = ax1.inset_axes([0.65, 0.65, 0.3, 0.3])
ax1_hist.hist(k_samples1, bins=20)
ax1_hist.set_title('k value distribution')
ax1_hist.set_xlim(0, 2)

ax2_hist = ax2.inset_axes([0.65, 0.65, 0.3, 0.3])
ax2_hist.hist(k_samples2, bins=20)
ax2_hist.set_title('k value distribution')
ax2_hist.set_xlim(0, 10)

plt.tight_layout()
plt.show()
```
